```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,85,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişimi")
#plt.show() ilk başta inline yazdığımız için bunu yazmaya gerek yok
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"y-*")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak , 2 kolon olacak ,şu an ikinci grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
# Artık kütüphaneden obje oluşturup onu kullanıcaz
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
# Arraylerin en başındaki iki tane float değer figurenin nerede duracağını belirliyor
# Mesela eksen2deki ilk değeri 0.1 yaparsam sola doğru kayar
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"y")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure , benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)
for eksen in benimEksenler :
     eksen.plot(numpyDizisi1,numpyDizisi2,"g")
     eksen.set_xlabel("X ekseni")
     eksen.set_ylabel("Y ekseni")

plt.tight_layout()
```

```python
yeniFigure = plt.figure()

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2 ")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label= "numpyDizisi ** 3")
yeniEksen.legend(loc=2)
```
